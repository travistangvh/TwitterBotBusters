```python
# recursively check if val is of type bigjson.obj.Object.
# If so, replace it with itself.items
def replace_bigjson(record):
    """Take in a list of tuple in the format of (key,value), and perform the following:
    1. Convert the list of tuple to a dictionary.
    2. For the dictionary, recursively perform the following:
    - if val is of type bigjson.obj.Object, replace it with itself.items
    - else, do nothing
    3. Once done, return the dictionary as is.
    """
    # Convert the list of tuple to a dictionary
    record = dict(record)
    # print(f"record: {record}")
    items = list(record.items())
    for key, val in items:
        # print(f"key, val: {key, val}")
        if isinstance(val, bigjson.obj.Object):
            # print(f"Selected key, val, val.items(): {key, val, val.items()}")
            record[key] = replace_bigjson(val)
            
        elif isinstance(val, bigjson.array.Array):
            # print(f"Entering an array: {key, val}")
            # print(val.to_python())
            record[key] = val.to_python()
    return record
```

```python
def get_data():# Sampling of tweets
    random.seed(42)
    lst = []

    tweet_0_list = []
    i = 0
    with open('./data/Twibot-22/tweet_0.json','rb') as f:
        records = bigjson.load(f)
        while True:
            try:
                if i > 1000:
                    break
                if records[i]['author_id'] in user_idx:
                    record_i = records[i].items() 
                    record_i = replace_bigjson(record_i)
                    tweet_0_list.append(record_i)
                    i=i+1
                    # print(f"Loop {i}")
                else:
                    i=i+1
                    # print(f"Loop {i}")
                    continue
                
            except:
                # print("Reached the end!")
                break

    with open('./data/Twibot-22-sampled-1000/tweet_0_time.json', 'a') as outfile:
        # Write final_str to file
        json.dump(tweet_0_list, outfile)
```
